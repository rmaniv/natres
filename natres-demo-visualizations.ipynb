```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
%matplotlib inline
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

## Reading all datasets

```python
total_natural_resources = pd.read_csv('/kaggle/input/natres/TOTAL_NATURAL_RESOURCE_RENTS.csv')
fossil_fuels = pd.read_csv('/kaggle/input/natres/FOSSIL_FUEL_RENTS.csv')
coal = pd.read_csv('/kaggle/input/natres/COAL_RENTS.csv')
oil = pd.read_csv('/kaggle/input/natres/OIL_RENTS.csv')
natural_gas = pd.read_csv('/kaggle/input/natres/NATURAL_GAS_RENTS.csv')
minerals = pd.read_csv('/kaggle/input/natres/MINERAL_RENTS.csv')
forest = pd.read_csv('/kaggle/input/natres/FOREST_RENTS.csv')
```

## Current INR per capita

```python
fig1 = px.line(total_natural_resources, x='Year', y='Current INR per capita', title='Total natural resource rents (current INR per capita)')
fig1.add_trace(go.Scatter(x = total_natural_resources['Year'], y = total_natural_resources['Current INR per capita'],
                         mode='markers',
                         marker=dict(color='red', size=6)))
fig1.update_layout(showlegend=False)
```
