```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df
```

```python
plt.plot(df['Year'],df['Coal rents (as % of GDP)'],'o-')
plt.xlabel('Year')
plt.ylabel('Coal rents (as % of GDP)')
plt.title('Coal rents (as a % of GDP) - World Bank')
plt.show()
```

```python
plt.plot(df['Year'],df['Mineral rents (as % of GDP)'],'o-')
plt.xlabel('Year')
plt.ylabel('Mineral rents (as % of GDP)')
plt.title('Mineral rents (as a % of GDP) - World Bank')
plt.show()
```

```python
plt.plot(df['Year'],df['Oil rents (as % of GDP)'],'o-')
plt.xlabel('Year')
plt.ylabel('Oil rents (as % of GDP)')
plt.title('Oil rents (as a % of GDP) - World Bank')
plt.show()
```

```python
plt.plot(df['Year'],df['Natural gas rents (as % of GDP)'],'o-')
plt.xlabel('Year')
plt.ylabel('Natural gas rents (as % of GDP)')
plt.title('Natural gas rents (as a % of GDP) - World Bank')
plt.show()
```

```python
plt.plot(df['Year'],df['Mineral rents (current INR)'],'o-')
plt.xlabel('Year')
plt.ylabel('Mineral rents (current INR)')
plt.title('Mineral rents (current INR) - World Bank')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Year'],df['Mineral rents (current INR per capita)'],'o-')
plt.xlabel('Year')
plt.ylabel('Mineral rents (as % of GDP)')
plt.title('Mineral rents (as a % of GDP) - World Bank')
for x, y in zip(df['Year'], df['Mineral rents (current INR per capita)']):
    if x > 2002:
        plt.annotate(f'{int(round(y))}', xy=(x, y), xytext=(5, 5), textcoords='offset points')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Year'],df['Mineral rents (2021 INR per capita)'],'o-')
plt.xlabel('Year')
plt.ylabel('Mineral rents (2021 INR per capita)')
plt.title('Mineral rents (2021 INR per capita) - World Bank')
for x, y in zip(df['Year'], df['Mineral rents (2021 INR per capita)']):
    if x > 2002:
        plt.annotate(f'{int(round(y))}', xy=(x, y), xytext=(5, 5), textcoords='offset points')
plt.show()
```
